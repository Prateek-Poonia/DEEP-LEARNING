```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
layers = 30
gradient = 1.0
sigmoid_grad = 0.25
gradients = []

for i in range(layers):

  gradient = gradient * sigmoid_grad
  gradients.append(gradient)


plt.figure(figsize=(10,5))
plt.plot(range(1,layers+1),gradients,marker="o", color = "orange")
plt.title("vanishing gradient across layers(Sigmoid activation)")
plt.xlabel("layer(from output to input)")
plt.ylabel("Gradient magnitude")
plt.grid(True)
plt.show()
```

```python
relu_gradients = [1.0 for _ in range(layers)]
plt.figure(figsize = (10,5))
plt.plot(range(1,layers+1),gradients,label="SIgmoid(vanishing)",marker="o", color = "red")

plt.plot(range(1,layers+1),relu_gradients,label = "Relu(stable)",marker="x", color = "green")
plt.title("vanishing gradient  vs (ReLU activation)")
plt.xlabel("layer(from output to input)")
plt.ylabel("Gradient magnitude")
plt.legend()
plt.grid(True)
plt.show()
```
