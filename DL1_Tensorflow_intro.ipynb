```python
import tensorflow as tf
import matplotlib.pyplot as plt
scaler_tensor = tf.constant(5)
print("Scaler Tesnor: ",scaler_tensor)

fig, ax = plt.subplots()
ax.text(0.5,0.5,str(scaler_tensor.numpy()),fontsize=14,ha='center',va='center')
ax.set_title('Scaler Tensor 0D tensor')
ax.axis('off')
plt.show()

```

```python
vector_tensor = tf.constant([1,2,3,4])
print("Vector Tensor: ",vector_tensor)
fig, ax = plt.subplots()
ax.plot(vector_tensor.numpy(),"o-",label="Values")
ax.text(0.5,0.5,str(scaler_tensor.numpy()),fontsize=14,ha='center',va='center')
ax.set_title('Vector Tensor 1D tensor')
ax.legend()

plt.show()

```

```python
matrix_tensor = tf.constant([[1,2],[3,4]])
print("Matrix tensor", matrix_tensor)

fig, ax = plt.subplots()
cax = ax.imshow(matrix_tensor.numpy(), cmap="Blues") # Use imshow for 2D data
fig.colorbar(cax)
ax.set_title('Matrix Tensor 2D tensor')
plt.show()
```

```python
tensor_3d = tf.constant([[[1,2],[3,4]],[[5,6],[7,8]]])
print("Tensor 3D", tensor_3d)

tensor_np = tensor_3d.numpy()
num_slices = tensor_np.shape[0]

fix,axs = plt.subplots(1,num_slices,figsize=(5*num_slices,4))
fig.suptitle('Tensor 3D')

if num_slices ==1:
  axs=[axs]

for i in range(num_slices):
  axs[i].imshow(tensor_np[i],cmap='Blues',interpolation="nearest")
  axs[i].set_title(f'Slice {i}')
  axs[i].axis("off")

plt.show()
```

```python
zeros_tensor = tf.zeros([3,3])
print("zeros tensor: ",zeros_tensor)

fig, ax = plt.subplots()
cax = ax.imshow(zeros_tensor.numpy(), cmap="Blues") # Use imshow for 2D data
fig.colorbar(cax)
ax.set_title('Zeros Tensor 2D tensor')
```
